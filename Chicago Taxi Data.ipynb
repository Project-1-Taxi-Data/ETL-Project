```python
timestamp = """SELECT
  trip_start_timestamp
FROM
  `bigquery-public-data.chicago_taxi_trips.taxi_trips`
WHERE
  dropoff_community_area IS NOT NULL
LIMIT
  10
        """
timestamp = chicago_taxi.query_to_pandas_safe(timestamp, max_gb_scanned=10)
timestamp
```

```python
querycount = """SELECT 
    count(1)
FROM
  `bigquery-public-data.chicago_taxi_trips.taxi_trips`
WHERE
  trip_seconds >= 600 AND
  trip_start_timestamp>= '2019-01-01 00:00:00+00:00' AND
      trip_start_timestamp < '2019-01-02 00:00:00+00:00' AND
      pickup_latitude IS NOT NULL
"""
taxi_data_count = chicago_taxi.query_to_pandas_safe(querycount, max_gb_scanned=68)
#taxi_data = response1.loc[response1["pickup_location"] != None]
taxi_data_count
```

```python
query1 = """SELECT 
    unique_key,
    trip_start_timestamp,
    ROUND(pickup_latitude, 2) AS lat,
    ROUND(pickup_longitude, 2) AS long,
    ROUND(trip_seconds / 60) AS duration_in_minutes,
    trip_miles,
    fare
FROM
  `bigquery-public-data.chicago_taxi_trips.taxi_trips`
WHERE
  trip_seconds >= 600 AND
  trip_start_timestamp>= '2019-01-01 00:00:00+00:00' AND
      trip_start_timestamp < '2019-01-07 00:00:00+00:00' AND
      pickup_latitude IS NOT NULL

"""
taxi_data_df = chicago_taxi.query_to_pandas_safe(query1, max_gb_scanned=68)
#taxi_data = response1.loc[response1["pickup_location"] != None]
#taxi_data.head(10)
```
